Code to generate Sierpinski's Triangle using Chaos Game

```python
plt.style.use('dark_background')
capture = 100
plt.rcParams['figure.figsize'] = (10.0, 10.0)
plt.scatter([0,2,4], [0,4,0])
bounds = ([0,0],[2,4],[4,0])
coordinates = [random.randrange(0,4) for i in range(2)]
#plt.text(2, -0.5, 'Created by Diksha Menghmalani, Shruti Dixit, Pallavi Thakare', style='italic',)
for i in range(n):
    plt.scatter(coordinates[0],coordinates[1], lw = '0',s = 20)
    plt.xlim([-0.5,4.5])
    plt.ylim([-0.5,4.5])
    plt.axis('off')
    plt.suptitle("Sierpinski's Triangle using Chaos Game!",fontsize = 35,color = 'white')
    if(i<100 and i%4 == 0):
        plt.text(0, -0.5, 'Created by Diksha Menghmalani, Shruti Dixit, Pallavi Thakare',fontsize = 15)
    plt.draw()
    random_boundry = np.array(random.choice(bounds))
    coordinates = coordinates+(0.5*(random_boundry-coordinates))
    if(i%capture == 0):
        plt.savefig(os.getcwd()+"/images/"+str(i)+"chaos_game"+str(n)+".png")
    if(i%500 == 0):
        plt.show()
```
